```python
import time
import random

class WaterPurificationSystem:
    def __init__(self, solar_panel_capacity):
        self.solar_panel_capacity = solar_panel_capacity
        self.solar_power_generated = 0
        self.water_tank = 0
        self.water_purity = 50  # Initial water purity level
        self.purification_rate = 0.1  # Simulated purification rate per second
        self.contaminant_generation_rate = 0.05  # Simulated contaminant generation rate per second

    def measure_solar_power(self):
        # Simulate solar power generation
        sunlight_intensity = random.uniform(0, 1)
        self.solar_power_generated = self.solar_panel_capacity * sunlight_intensity

    def purify_water(self):
        # Simulate water purification using solar power
        purification_capacity = self.solar_power_generated * self.purification_rate
        self.water_tank += purification_capacity

        # Simulate contaminant generation
        contaminant_generation = self.water_tank * self.contaminant_generation_rate
        self.water_tank -= contaminant_generation

        # Update water purity based on the balance between purification and contaminant generation
        self.water_purity = max(0, min(100, self.water_purity + purification_capacity - contaminant_generation))

    def get_water_level(self):
        return self.water_tank

    def get_water_purity(self):
        return self.water_purity

    def simulate_time_step(self):
        # Simulate a time step (e.g., 1 second)
        self.measure_solar_power()
        self.purify_water()

# Initialize the water purification system with a solar panel capacity of 100 units
water_system = WaterPurificationSystem(solar_panel_capacity=100)

# Simulate continuous operation and water purification
for _ in range(10):
    water_system.simulate_time_step()

    # Print the results
    print(f"Solar Power Generated: {water_system.solar_power_generated} units")
    print(f"Water Level: {water_system.get_water_level()} units")
    print(f"Water Purity: {water_system.get_water_purity()}%")
    print("------")

    time.sleep(1)  # Simulate a 1-second time step
```
